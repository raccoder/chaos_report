```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
fig, ax = plt.subplots()
ax.set_xlim(0,0.5)
ax.set_ylim(0,1)
line, = ax.plot([], [], 'b-')

Xn = [0.19]  # åˆæœŸå€¤
yn = [0.1]   # åˆæœŸå€¤
def xn1(xn, yn):
    return 1.9 * xn * (1 - xn) - xn * yn
def yn1(xn, yn):
    return 1 / 0.25 * xn * yn
def init():
    line.set_data([], [])
    return line,
def animate(i):
    global Xn, yn
    Xn.append(xn1(Xn[-1], yn[-1]))
    yn.append(yn1(Xn[-1], yn[-1]))
    line.set_data(Xn, yn)
    return line,
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=200, blit=True)

plt.show()
```
